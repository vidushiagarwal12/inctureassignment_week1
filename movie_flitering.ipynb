```python
import pandas as pd
```

```python
# Read CSV file
df = pd.read_csv(r"C:\week1_assignment\movie.csv")
```

```python
print("\n                          COMPLETE MOVIE DATASET           \n")
print(df)
```

```python
def filter_movies(genre, rating):

    # Filter condition
    filtered_df = df[
        (df["Genre"].str.lower() == genre.lower()) &
        (df["Rating"] >= rating)
    ]

    # Return filtered dataframe
    return filtered_df
```

```python
genre_input = input("\nEnter Genre: ")
rating_input = float(input("Enter Minimum Rating: "))
```

```python
result = filter_movies(genre_input, rating_input)
print("\n                               FILTERED MOVIES                   \n")
if not result.empty:
    print(result.to_string(index=False))
else:
    print("No movies found matching the given criteria.")
```

```python
print("\n========== SUMMARY ==========")
print("Total Movies in Dataset :", len(df))
print("Filtered Movies Count   :", len(result))
```
